Podemos revisar el coste en los comentarios del código anterior y si ponemos toda la formula de recursividad
$$f(N)=f(N−1)+1+f(N−1)$$
$$f(N)=2f(N−1)+1$$
Paso 1 sustituimos para encontrar la recursividad f(N-1)
$$f(N)=2[2f(N−2)+1]+1$$
$$f(N)=4f(N−2)+2+1$$
paso 2 sustituimos por $f(N-2)=2f(N−3)+1$
$$f(N)=4[2f(N−3)+1]+2+1$$
$$f(N)=8f(N−3)+4+2+1$$
Podemos reescribir la ecuación como:
$$f(N)=2^3f(N−3)+2^3+2^2+2^0$$
podenos observar un patron el cual podemos generalizarlo de la siguiente forma:
$$f(N)=2^{N-1}f(1)+2^{N-2}+...+2^2+2^0$$
donde $f(1)=1$
$$f(N)=2^{N-1}+2^{N-2}+...+2^2+2^0$$

Con lo anterior podemos verificar que la complejidad es $O(2^N)$ una complejidad exponencial

```python
movimientos = [0]
def Fibonacci_Memo(N, memoria={}):
    # 1. ¿Ya calculamos este valor? ¡Devuélvelo sin calcular nada!
    if N in memoria:
        return memoria[N]
        
    # 2. Casos base (igual que tu código)
    if N < 2:
        return 1
        
    # 3. Calculamos, pero GUARDAMOS el resultado antes de devolverlo
    else:
        resultado = Fibonacci_Memo(N-1, memoria) + Fibonacci_Memo(N-2, memoria)
        movimientos[0] +=1
        memoria[N] = resultado # Guardamos en el "cuaderno"
        return resultado

print(Fibonacci_Memo(5, {}))
print(f"total movimientos Fibonacci_Memo = {movimientos}") 
movimientos = [0]
def Fibonacci(N:int):
  if N < 2:
    return 1 ## coste 1
  else:
    movimientos[0] +=1
    return Fibonacci(N-1)+Fibonacci(N-2) ##coste f(N-1) + f(N-2)

print(Fibonacci(5))
print(f"total movimientos Fibonacci = {movimientos}") 
```

##Devolución de cambio por técnica voraz

```python
def cambio_monedas(N, SM):
  SOLUCION = [0]*len(SM)   #SOLUCION = [0,0,0,0,..]
  ValorAcumulado = 0

  for i,valor in enumerate(SM):
    monedas =  (N-ValorAcumulado)//valor
    SOLUCION[i] = monedas
    ValorAcumulado = ValorAcumulado + monedas*valor

    if ValorAcumulado == N:
      return SOLUCION


cambio_monedas(15,[25,10,5,1])
```

```python
def escribe(S):
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")


def es_prometedora(SOLUCION,etapa):
  print(SOLUCION)
  #Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    print("El valor " + str(SOLUCION[i]) + " está " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True



def reinas(N, solucion=[], etapa=0):
  if len(solucion) == 0:
      solucion=[0 for i in range(N)]

  for i in range(1, N+1):
    solucion[etapa] = i

    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
        #escribe(solucion)
        print()
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

    solucion[etapa] = 0

reinas(4)
```

Revisamos que tan bueno es ejecutando el programa

```python
import matplotlib.pyplot as plt
movimientos = [0]
def es_prometedora(SOLUCION, etapa):
    # Contamos las iteraciones dentro de la validación
    for i in range(etapa + 1):
        movimientos[0] += 1
        
        # Nota: SOLUCION.count() internamente también recorre la lista O(N)
        if SOLUCION.count(SOLUCION[i]) > 1:
            return False

        # Verifica las diagonales
        for j in range(i + 1, etapa + 1):
            movimientos[0] += 1
            if abs(i - j) == abs(SOLUCION[i] - SOLUCION[j]): 
                return False
                
    return True

def reinas(N, solucion=None, etapa=0):
    # Inicializamos la solución vacía en la primera llamada
    if solucion is None:
        solucion = [0 for _ in range(N)]

    for i in range(1, N + 1):
        solucion[etapa] = i
        movimientos[0] += 1

        if es_prometedora(solucion, etapa):
            if etapa == N - 1:
                # Encontramos una solución. (Omitimos el print para no saturar)
                pass
            else:
                reinas(N, solucion, etapa + 1)

        # Deshacemos el movimiento (Backtracking)
        solucion[etapa] = 0

# --- LÓGICA DE GRAFICACIÓN ---
N = []
operaciones_totales = []

# Probamos tamaños de tablero desde N=4 hasta N=10
# (Más allá de 10 o 11 puede tardar varios minutos u horas)
for n in range(1, 11):
    movimientos = [0]
    reinas(n, solucion=None, etapa=0)
    
    N.append(n)
    operaciones_totales.append(movimientos[0])
    print(f"Para N={n}, el algoritmo realizó {movimientos[0]} operaciones.")

# Creación de la gráfica
plt.figure(figsize=(8, 5))
plt.plot(N, operaciones_totales, marker='o', color='purple', linewidth=2)
plt.title('Complejidad Algoritmo N-Reinas (Backtracking)')
plt.xlabel('N (Tamaño del Tablero)')
plt.ylabel('Operaciones Totales (escala normal)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

como observamos este algoritmo crece de forma exponencial $O(n!)$

## Viaje por el rio. Programación dinámica

```python
TARIFAS = [
[0,5,4,3,999,999,999],
[999,0,999,2,3,999,11],
[999,999, 0,1,999,4,10],
[999,999,999, 0,5,6,9],
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]



################################################################
def Precios(TARIFAS):
################################################################
  #Total de Nodos
  N = len(TARIFAS[0])

  #Inicialización de la tabla de precios
  PRECIOS = [ [9999]*N for i in [9999]*N]
  RUTA = [ [""]*N for i in [""]*N]

  for i in range(0,N-1):
    RUTA[i][i] = i             #Para ir de i a i se "pasa por i"
    PRECIOS[i][i] = 0          #Para ir de i a i se se paga 0
    for j in range(i+1, N):
      MIN = TARIFAS[i][j]
      RUTA[i][j] = i

      for k in range(i, j):
        if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
            MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
            RUTA[i][j] = k          #Anota que para ir de i a j hay que pasar por k
        PRECIOS[i][j] = MIN

  return PRECIOS,RUTA
################################################################


PRECIOS,RUTA = Precios(TARIFAS)
#print(PRECIOS[0][6])

print("PRECIOS")
for i in range(len(TARIFAS)):
  print(PRECIOS[i])

print("\nRUTA")
for i in range(len(TARIFAS)):
  print(RUTA[i])

#Determinar la ruta con Recursividad
def calcular_ruta(RUTA, desde, hasta):
  if desde == hasta:
    #print("Ir a :" + str(desde))
    return ""
  else:
    return str(calcular_ruta( RUTA, desde, RUTA[desde][hasta])) +  \
                ',' + \
                str(RUTA[desde][hasta] \
              )

print("\nLa ruta es:")
calcular_ruta(RUTA, 0,6)
```

```python
def cambio_monedas_PD(N, SM):

    #Inicialización de la tabla
    dp = [float("inf")]*(N+1)
    dp[0] = 0
    monedas = []

    for i in range(1, N+1):
        for moneda in SM:
            if i >= moneda:
                if dp[i] > dp[i-moneda]+1:
                    dp[i] = dp[i-moneda]+1
                    print(f"I= {i}")
                    print(f"moneda= {moneda}")
                    print(f"dp= {dp}")
                else:
                    dp[i] = dp[i]

    if dp[N] == float("inf"):
        print("No se encontró una solución")
        return -1
    else:
        print(dp)
        print(dp[N])



cambio_monedas_PD(15,[25,10,5,1])
cambio_monedas_PD(6, [1, 3, 4])
```
